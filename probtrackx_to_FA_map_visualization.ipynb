```python
import os
import re
import nibabel as nib
import nilearn
%matplotlib inline
import matplotlib.pyplot as plt
from nilearn import plotting, image
import matplotlib
import nipype.interfaces.fsl as fsl
```

```python
dataLocation = '/Volumes/CCNC_3T_2/kcho/ccnc/GHR_project'
```

```python
with open('subjects.txt','r') as f:
    subjects = f.read().split('\n')
```

```python
for subject in subjects:
    for side in ['lh','rh']:
        fdtDir = os.path.join(dataLocation,
                                  subject,
                                  'tractography',
                                  side)
        fdtLoc = os.path.join(fdtDir, 'fdt_paths.nii.gz')
        
        try:
            with open(os.path.join(fdtDir,'waytotal'),'r') as f:
                waytotal = int(f.read())
        except:
            waytotal = 0

        try:
            print subject
            newFdt = os.path.join(fdtDir, 'fdt_paths_norm.nii.gz')
            subjectDir = os.path.join(dataLocation, subject)
            T1Loc = os.path.join(dataLocation, subject, 'FREESURFER/mri/brain.nii.gz')
            wmMask = os.path.join(dataLocation, subject, 'ROI/{0}_wm_mask.nii.gz'.format(side))
            divider = fsl.ImageMaths(in_file = fdtLoc,
                                     op_string = '-div {0} -thrp 2 -bin -mul {1}'.format(waytotal, wmMask),
                                     out_file = newFdt)
            print divider.cmdline
            divider.run()
            plotting.plot_stat_map(newFdt,
                              bg_img = T1Loc,
        #                      cmap = matplotlib.colors.ListedColormap('red'),
#                              threshold=.2,
                              title=subject,
                              output_file=subject+'_{0}.png'.format(side))
        except:
            print subject, 'has errors'

```
